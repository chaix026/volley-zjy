# Volleyball statistics analysis

```python
# Scrape data from volley.ru
import requests
from bs4 import BeautifulSoup
import pandas as pd
from googletrans import Translator
import csv
```

```python
# URL of the webpage containing the table
# url = 'https://volley.ru/calendar/01H6XABKE6DZMV0NXGVB2Z0CK5/statistics'
url = 'https://volleyservice.ru/index.php?option=com_volleyplayers&task=statistic&act=main_stat'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table in the HTML using its tag (usually <table>)
table = soup.find('table')

# Extract data from the table
table_rows = table.find_all('tr')

# Loop through each row in the table
for row in table_rows:
    # Extract data from each cell in the row
    cells = row.find_all('td')
    row_data = [cell.text.strip() for cell in cells]
    print(row_data)  # Or do whatever you want with the data
```

```python
# Initialize the translator
translator = Translator()

# Text to translate
text_to_translate = "Привет, как дела?"

# Translate from Russian to English
translated_text = translator.translate(text_to_translate, src='ru', dest='en')

# Output the translated text
print(translated_text.text)

# Function to translate text using Google Translate API
def translate_text(text):
    translation = translator.translate(text, src='ru', dest='en')
    return translation.text
```

```python
# Save data into a dataframe
players_stats = table_rows[4:]

# Loop through each row in the table
rows = []
for row in players_stats:
    # Extract data from each cell in the row
    cells = row.find_all('td')
    row_data = [cell.text.strip() for cell in cells]
    # print(row_data)  # Or do whatever you want with the data
    rows.append(row_data)

df =  pd.DataFrame(rows)
df.head()
```

```python
# Translate player names
df[1] = df[1].apply(translate_text)
```

```python
df.columns=['No.', 'Player', 'Games/Sets', 
            'Points_All', 'Points_Avg', 'Points_Net',
            'Serve_All', 'Serve_Points', 'Serve_SetAvg',
            'Reception_All', 'Reception_Pos%', 'Recept_Eff%', 
            'Attack_All', 'Attack_Points', 'Attack_Eff%',
            'Block_Points', 'Block_SetAvg',
            'Error_Serve', 'Error_Recept', 'Error_Atk', 'Error_All']

# Display the DataFrame
print(df)
```
